```python
import pandas as pd

df = pd.read_csv("/workspaces/data_3500_in_class_code_FA_2025/Lecture_code/NVDA.csv")
df
```

```python
# clean the string
df["Close/Last"] = df["Close/Last"].replace("[$]", "", regex=True).astype(float)
# print(df.info())

df = df.iloc[::-1].reset_index(drop=True)
df
```

```python
# moving avg

df["Moving Avg"] = df["Close/Last"].shift(1).rolling(window=5).mean()
df.head(10)
```

```python
buy             = 0
first_buy       = 0
total_profit    = 0


for i in range(len(df)):
    current_price = df.loc[i, "Close/Last"]
    average_price = df.loc[i, "Moving Avg"]

    if current_price < (average_price * 0.98) and buy == 0:
        buy = current_price
        print("Buying at: \t", buy)
        if first_buy == 0:
            first_buy = current_price
    elif current_price > (average_price * 1.02) and buy != 0:
        trade_profit = current_price - buy
        buy = 0
        print("Selling at: \t", current_price)
        print("Trade Profit: \t", trade_profit)
        total_profit += trade_profit
```

```python
# output results

print("------------------------------------")
print("Total Profit: \t", total_profit)
print("First Buy: \t", first_buy)
print("% Return: \t", (total_profit/first_buy) * 100)
```
