```python
# importing data and library
import pandas as pd

df = pd.read_csv("/workspaces/data_3500_in_class_code_FA_2025/Lecture_code/titanic.csv")
df.head()
```

```python
# info
df.info()
```

```python
# identifying nulls
df.isnull().sum()
```

```python
# filling nulls and dropping columns
df["Age"].fillna(df["Age"].median(), inplace=True)
# df["Age"].head(7)
# df.dropna(subset=["Cabin"], inplace=True)
df.head(10)
```

```python
# rename column
df = df.rename(columns={"Pclass":"Class"})
df.rename(columns={"Parch":"Parent/Child"}, inplace=True)
df.rename(columns={"SibSp":"Sibling/Spouse"}, inplace=True)
df.head()
```

```python
# change data types
df["Survived"] = df["Survived"].astype(bool)
df.head()
```

```python
# create new column
df["Family Size"] = df["Sibling/Spouse"] + df["Parent/Child"] + 1
df.head()
```

```python
# grouping
df.groupby("Sex")["Survived"].mean()
df.groupby(["Class", "Sex"])["Survived"].mean()
```

```python
# sort and aggregate
df.groupby("Class")["Fare"].agg(["mean", "median", "count"])
```

```python
# visualize
df["Age"].hist(bins=20)
df["Sex"].value_counts().plot(kind="bar")
```
